## Error

Error between numerical and analytical

#### Absolute error

```python
x = []
AbsY10 = []
for line in open('AbsError10.txt', 'r'):
    lines = [i for i in line.split()]
    x.append(float(lines[0]))
    AbsY10.append(float(lines[1]))
      

AbsY100 = []
x100 = []
for line in open('AbsError100.txt', 'r'):
    lines = [i for i in line.split()]
    x100.append(float(lines[0]))
    AbsY100.append(float(lines[1]))

x1000 = []
AbsY1000 = []
for line in open('AbsError1000.txt', 'r'):
    lines = [i for i in line.split()]
    x1000.append(float(lines[0]))
    AbsY1000.append(float(lines[1]))

plt.figure(figsize=(17, 15))

plt.title('Absolute error for different number of steps', fontsize=15)
plt.xlabel('x', fontsize=15)

plt.plot(x, AbsY10)
plt.plot(x100, AbsY100)
plt.plot(x1000, AbsY1000)
plt.legend(['10 steps', '100 steps', '1000 steps'], loc = 'upper left', fontsize=15)
plt.show()
```

#### Relative error

```python
x = []
RelY10 = []
for line in open('RelError10.txt', 'r'):
    lines = [i for i in line.split()]
    x.append(float(lines[0]))
    RelY10.append(float(lines[1]))
      

RelY100 = []
x100 = []
for line in open('RelError100.txt', 'r'):
    lines = [i for i in line.split()]
    x100.append(float(lines[0]))
    RelY100.append(float(lines[1]))

x1000 = []
RelY1000 = []
for line in open('RelError1000.txt', 'r'):
    lines = [i for i in line.split()]
    x1000.append(float(lines[0]))
    RelY1000.append(float(lines[1]))

plt.figure(figsize=(17, 10))

plt.title('Relative error for different number of steps', fontsize=15)
plt.xlabel('x', fontsize=15)

plt.plot(x, RelY10)
plt.plot(x100, RelY100)
plt.plot(x1000, RelY1000)
plt.legend(['10 steps', '100 steps', '1000 steps'], loc = 'upper left', fontsize=15)
plt.show()
```

## Elapsed time 

```python
import numpy as np 

hard_time10 = 0.000016
easy_time10 = 0.000006

hard_time100 = 0.000042
easy_time100 = 0.000027

hard_time1000 = 0.000250
easy_time1000 = 0.000180

hard_time10000 = 0.002761
easy_time10000 = 0.002075

hard_time100000 = 0.018884
easy_time100000 = 0.011080

hard_time1000000 = 0.096150
easy_time1000000 = 0.069940

hard_times = [hard_time10, hard_time100, hard_time1000, hard_time10000, hard_time100000, hard_time1000000]
easy_times = [easy_time10, easy_time100, easy_time1000, easy_time10000, easy_time100000, easy_time1000000]

x = np.linspace(1, 6, 6)
plt.figure(figsize=(12, 10))

plt.title('Run time for the special and general algorithm', fontsize=15)
plt.xlabel('steps to the power of x', fontsize=15)
plt.ylabel('time [s]', fontsize=15)

plt.plot(x, hard_times,label="General alogorithm")
plt.plot(x, easy_times, label="Special algorithm")
plt.legend(loc="upper left", fontsize=15)
plt.show()
```
